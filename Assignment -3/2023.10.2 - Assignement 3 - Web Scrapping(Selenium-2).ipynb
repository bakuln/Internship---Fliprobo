##### Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: 
"Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
driver = webdriver.Chrome()

driver.get('https://www.amazon.in/')

item = driver.find_element(By.XPATH,'//input[@type="text"]')
item.send_keys(inp)

search = driver.find_element(By.XPATH,'//input[@id="nav-search-submit-button"]')
search.click()

URL=[]

start=0
end=3
for page in range(start, end):
    URL_tags = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in URL_tags:
        URL.append(i.get_attribute('href'))
    
    time.sleep(3)    
    
    try:
        nxt=driver.find_element(By.XPATH,'//span[@class="s-pagination-strip"]/a[4]')
        nxt.click()
    
    except:
        nxt=driver.find_element(By.XPATH,'//span[@class="s-pagination-strip"]/a[3]')
        nxt.click()

#variables as required
brnm=[]
prnm=[]
price=[]
retex=[]
exdel=[]
avai=[]

for i in URL:
    driver.get(i)
#    time.sleep(2)
    try:
        brnm_tags=driver.find_element(By.XPATH, '//span[@class="a-size-base po-break-word"]')
        brnm.append(brnm_tags.text)
    except NoSuchElementException :
        brnm.append(' - ')
    
    try:
        prnm_tags=driver.find_element(By.XPATH, '//span[@class="a-size-large product-title-word-break"]')
        prnm.append(prnm_tags.text)
    except NoSuchElementException :
        prnm.append(' - ')
    
    try:
        price_tags=driver.find_element(By.XPATH, '//span[@class="a-price-whole"]')
        price.append(price_tags.text)
    except NoSuchElementException :
        price.append(' - ')
        
    try:
        retex_tags=driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span')
        retex.append(retex_tags.text)
    except NoSuchElementException :
        retex.append(' - ')

    try:
        exdel_tags=driver.find_element(By.XPATH, '//span[@class="a-text-bold"]')
        exdel.append(exdel_tags.text)
    except NoSuchElementException :
        exdel.append(' - ')
        
    try:
        avai_tags=driver.find_element(By.XPATH, '//span[@class="a-size-medium a-color-success"]')
        avai.append(avai_tags.text)
    except NoSuchElementException :
        avai.append(' - ')

print(len(brnm),len(prnm),len(price),len(retex),len(exdel),len(avai))

guitardf=pd.DataFrame({'Brand Name':brnm, 'Product Name':prnm, 'Price':price, 'Return/Exchange':retex, 'Expected Delivery':exdel, 'Availability':avai,'Page URL':URL})

guitardf.to_csv('Q2.CSV',index=False)

guitardf
```

##### Q4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped:
“Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV. 

```python
driver = webdriver.Chrome()

driver.get('https://www.flipkart.com/')

close = driver.find_element(By.XPATH, '/html/body/div[3]/div/span')
close.click()

item = driver.find_element(By.XPATH,'//input[@class="Pke_EE"]')
item.send_keys('Smartphone')

search = driver.find_element(By.XPATH,'//button[@class="_2iLD__"]')
search.click()

URL=[]
URL_tags = driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in URL_tags:
    URL.append(i.get_attribute('href'))
    
brnm=[] #Brnad name
spnm=[] #Smartphone Name
col=[] #Color
ram=[] #RAM
rom=[] #ROM
pcam=[] #Primary Camera
scam=[] #Secondary Camera
displ=[] #Display Size
batcap=[] # Battery Capacity
price=[] # Price

for i in URL:
    driver.get(i)
#    time.sleep(2)

# Smartphone Name
    try:
        spnm_tags=driver.find_element(By.XPATH, '//span[@class="B_NuCI"]')
        spnm.append(spnm_tags.text)
    except NoSuchElementException :
        spnm.append(' - ')
        
# Color        
    try:
        col_tags=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div/div[1]/table/tbody/tr[4]/td[2]')
        col.append(col_tags.text)
    except NoSuchElementException :
        col.append(' - ')
        
# RAM & ROM
    try:
        raro_tags = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]')
        rarot=raro_tags.text.replace(' | ',",")
        tem = rarot.split(",")
        ram_tag = tem[0]
        rom_tag = tem[1]
        ram.append(ram_tag)
        rom.append(rom_tag)
    except NoSuchElementException :
        ram.append(' - ')
        rom.append(' - ')

# Primary & Secondary Cameras
    try:
        cam_tags = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[3]')
        camt=cam_tags.text.replace(' | ',",")
        tem = camt.split(",")
        pcam_tag = tem[0]
        scam_tag = tem[1]
        pcam.append(pcam_tag)
        scam.append(scam_tag)
    except NoSuchElementException :
        pcam.append(' - ')
        scam.append(' - ')
        
# Price        
    try:
        price_tags=driver.find_element(By.XPATH, '//div[@class="_30jeq3 _16Jk6d"]')
        price.append(price_tags.text)
    except NoSuchElementException :
        price.append(' - ')
        
# Display Size
    try:
        displ_tags=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[2]')
        displ.append(displ_tags.text)
    except NoSuchElementException :
        displ.append(' - ')
        
#Battery Capacity
    try:
        batcap_tags=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[4]')
        batcap.append(batcap_tags.text.replace(' Lithium Ion Battery',""))
    except NoSuchElementException :
        batcap.append(' - ')
        
# Brand Name
    try:
        knmr = driver.find_element(By.XPATH, '//span[@class="_3IIjhd"]')
        knmr.click()
        brnm_tags=driver.find_element(By.XPATH, '//section[@class="_1-ermZ"]/h4')
        brnm.append(brnm_tags.text)
    except NoSuchElementException :
        brnm.append(' - ')

driver.close()


df=pd.DataFrame({'Brand Name':brnm, 'Smartphone Name':spnm, 'Color':col, 'RAM':ram, 'Storage(ROM)':rom, 'Primary Camera':pcam,'Secondary Camera':scam, 'Display':displ,'Battery':batcap,'Price':price, "Product URL":URL})

df.to_csv('Q4.CSV',index=False)

df
```

##### Q7.Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped:
“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
driver = webdriver.Chrome()

driver.get('https://www.forbes.com/billionaires/')

rank=[]
name=[]
ntwr=[]
age=[]
country=[]
sourc=[]
indus=[]

start = 0
end = 13
for page in range(start,end):
    rank_tags = driver.find_element(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]/div[1]/div')
    try:
        rank.append(rank_tags.text)
    except NoSuchElementException :
        rank.append(' - ')

    name_tags = driver.find_element(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]/div[2]/div')
    try:
        name.append(name_tags.text)
    except NoSuchElementException :
        name.append(' - ')
    
    ntwr_tags = driver.find_element(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]/div[3]/div')
    try:
        ntwr.append(ntwr_tags.text.replace("$",""))
    except NoSuchElementException :
        name.append(' - ')
    
    age_tags = driver.find_element(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]/div[4]/div')
    try:
        age.append(age_tags.text)
    except NoSuchElementException :
        age.append(' - ')

    country_tags = driver.find_element(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]/div[5]')
    try:
        country.append(country_tags.text)
    except NoSuchElementException :
        country.append(' - ')

    sourc_tags = driver.find_element(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]/div[6]/div')
    try:
        sourc.append(sourc_tags.text)
    except NoSuchElementException :
        sourc.append(' - ')                                    

    indus_tags = driver.find_element(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]/div[7]/div')
    try:
        indus.append(indus_tags.text)
    except NoSuchElementException :
        indus.append(' - ')

    rank_tags = driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km"]/div[1]/div')
    try:
        for i in rank_tags:
            rank.append(i.text)
    except NoSuchElementException :
        rank.append(' - ')

    name_tags = driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km"]/div[2]/div')
    try:
        for i in name_tags:
            name.append(i.text)
    except NoSuchElementException :
        name.append(' - ')
    
    ntwr_tags = driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km"]/div[3]/div')
    try:
        for i in ntwr_tags:
            ntwr.append(i.text.replace("$",""))
    except NoSuchElementException :
        ntwr.append(' - ')
    
    age_tags = driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km"]/div[4]/div')
    try:
        for i in age_tags:
            age.append(i.text)
    except NoSuchElementException :
        age.append(' - ')

    country_tags = driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km"]/div[5]')
    try:
        for i in country_tags:
            country.append(i.text)
    except NoSuchElementException :
        country.append(' - ')

    sourc_tags = driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km"]/div[6]/div')
    try:
        for i in sourc_tags:
            sourc.append(i.text)
    except NoSuchElementException :
        sourc.append(' - ')                                    

    indus_tags = driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km"]/div[7]/div')
    try:
        for i in indus_tags:
            indus.append(i.text)
    except NoSuchElementException :
        indus.append(' - ')
    
    nxt=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div[2]/div[2]/div[2]/div[1]/button[2]/div/span[1]')
    nxt.click()
    
        
df=pd.DataFrame({'Rank':rank, 'Name':name, 'Net Worth(in $)':ntwr, 'Age':age, 'Citizenship':country, 'Source':sourc, 'Industry':indus})
df.to_csv('Q7.CSV',index=False)
df
```

##### Q8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.
