##### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
# Start Chrome
driver = webdriver.Chrome()

# Open Link
driver.get("https://www.shine.com/")

# Stop for Pop-Up, not required for actual scrping of 01 Page.
#time.sleep(15)

# Close Pop-Up, not required for actual scraping of 01 Page.
#close_popup=driver.find_element(By.CLASS_NAME,"btn-close")
#close_popup.click()

# Click on Search Bar
searchbar=driver.find_element(By.CLASS_NAME,"searchBarInput")
searchbar.click()

# Wait for Search bar to be active
time.sleep(2)

#Search for Designation
designation=driver.find_element(By.CLASS_NAME,"form-control")
designation.send_keys('Data Analyst')

#Search for Location
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')

#Click Search
search=driver.find_element(By.XPATH,'//button[@class=" btn btn-secondary undefined"]')
search.click()

# Stop for Pop-Up, not required for actual scraping of 01 Page.
#time.sleep(15)

# Close Pop-Up, not required for actual scraping of 01 Page.
#close_popup=driver.find_element(By.XPATH,'//button[@class="btn-close register_close"]')
#close_popup.click()

# search variables
job_title=[]
job_location=[]
company_name=[]
experience_required=[]

title_tags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
    
location_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags[0:10]:
    location=i.text.replace("\n"," ")
    job_location.append(location)
    
company_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)

experience_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags[0:10]:
    experience=i.text
    experience_required.append(experience)

df = pd.DataFrame({'Job Title':job_title, 'Job Location':job_location, 'Company Name':company_name, 'Experience Reqd.':experience_required})
df
```

##### Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
6. Brand
7. Product Description
8. Price

```python
driver = webdriver.Chrome()

driver.get('https://www.flipkart.com/')

time.sleep(3) #wait for Pop-Up

popup = driver.find_element(By.XPATH,'//span[@role="button"]') #close Pop-Up
popup.click()

search = driver.find_element(By.XPATH,'//input[@class="Pke_EE"]')
search.send_keys('sunglasses')

search_item = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/button')
search_item.click()

brand=[]
prod_disc=[]
price=[]
discount=[]

start = 0
end = 3
for page in range(start,end):
    brand_tags = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brand_tags[0:100]:
        brand.append(i.text)
        
    prod = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in prod[0:100]:
        prod_disc.append(i.text)
    
    price_tags = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in price_tags[0:100]:
        price.append(i.text)
        
    discou = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in discou[0:100]:
        discount.append(i.text)
        
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)


df=pd.DataFrame({'Brand':brand[0:100],'Product Description':prod_disc[0:100],'Price':price[0:100],'Discount':discount[0:100]})
df
```

##### Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone.

##### Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.

```python
driver = webdriver.Chrome()

driver.get('https://www.azquotes.com/')

sh = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/a[1]')
sh.click()

tquote = driver.find_element(By.XPATH,'//a[@href="/top_quotes.html"]')
tquote.click()

quote=[]
auth=[]
typ=[]

start=0
end=10
for page in range(start,end):
    qu = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in qu[0:1000]:
        quote.append(i.text)
    
    au = driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in au[0:1000]:
        auth.append(i.text)
        
    ty = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in ty[0:1000]:
        typ.append(i.text)
        

df=pd.DataFrame({'Quote':quote,'Author':auth,'Type of Quote':typ})
df
```

##### Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/.

##### Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/
